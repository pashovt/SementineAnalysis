<p style='text-align: justify;'>
The prediction in the kNN algorithm is based around the selected k value where k ∈ $\{1, 3, 5,\ldots, n+2\}$ and k is the number of nearest neighbours that will be used as reference. A odd number is selected for the k value so that there won't be a equal number of niegbour from different clases. k also needs to be ≠ to the number of classes. A visual representation of the algorithm is shown by drowning a circle as indicate on the figure below. A example of k with a value of 3 is shown. It can be seen that with a value of 3 for k the point of interest being the Green has 2 Red neighbours and 1 Blue neigbour. As such the Green point has a higher probability that is from the Red class. In kNN the number of preveiling number of neighbours from the same class in the corresponding k-sized neibourhood is the new classification of the point of interest. In this case the Green point is classed to the Red class. If the k value is specified to be 5 the point of interest will be classified in the other class being the Blue. A proper method of finding the k value and not setting it as a default is to use a cross validation techniques and test several value of k. From those test the most sutable value for the data is chosen and used in the kNN algorithm
</p>

<p style='text-align: justify;'>
The algorithm uses the least distance measure to class the number of point from the different classes in the specified neihbourhood. The used approch of the least distance measurement has been chosen to be Euclidea distance being the direct distance between two point and is calculated using the $\sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}$ formula where $x_{1,2}$ any $y_{1,2}$ are the x and y condinates of the 2 computed points.

### Step 0 - Read data

```python
import csv
with open('iris.txt') as csvfile:
    lines = csv.reader(csvfile)
    for row in lines:
        print(', '.join(row))
```

### Step 1 - Split the data randomly into training and testing

```python
import random

def loadingdataset(Filename, split, training=[], test=[]):
    with open(Filename, 'r') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        for x in range(1,len(dataset)-1):
            for y in range(4):
                dataset[x][y] = float(dataset[x][y])
            if random.random() < split:
                training.append(dataset[x])
            else:
                test.append(dataset[x])
```

```python
training = []
test = []

loadingdataset(r'iris.txt', 0.66, training, test)
print('Train: ' + repr(len(training)))
print('Test: ' + repr(len(test)))
```

### Step 2 - Simularity

```python
# Calculates the distance between first and second instances
import math

def euclidean(firstInstance, secondInstance, ln):
    gapbetweenPoints = 0
    for X in range(ln):
        gapbetweenPoints += pow(firstInstance[X] - secondInstance[X], 2)
    return math.sqrt(gapbetweenPoints)
```

```python
DATA1 = [3, 1, 2, 'a']
DATA2 = [4, 3, 4, 'b']
Distance = euclidean(DATA1, DATA2, 3)
print('Distance is' + repr(Distance))
```

### Step 3 - Find the k-nearest neighbours

```python
import operator

def Neighbors(training, test, k):
    Distance = []
    ln = len(test)-1
    for X in range(len(training)):
        dist = euclidean(test, training[k], ln)
        Distance.append((training[X], dist))
    Distance.sort(key=operator.itemgetter(1))
    neighbors = []
    for X in range(k):
        neighbors.append(Distance[X][0])
    return neighbors
```

```python
Training = [[3, 1, 2, 'a'], 
    [4, 3, 4, 'b']]
test = [6,6,6]
k = 1
neighbors = Neighbors(Training, test, k)
print(neighbors)
```

```python

def Response(neighbors):
    Votes = {}
    for X in range(len(neighbors)):
        resp = neighbors[X][-1]
        if resp in Votes:
            Votes[resp] += 1
        else:
            Votes[resp] = 1
    sorter = sorted(Votes.items(), key=operator.itemgetter(1), reverse=True)
    return sorter[0][0]
```

```python
neighbors = [[4,2,3,'a'], [3,2,2,'a'], [5,3,6,'b']]
resp = Response(neighbors)
print(resp)
```

```python
def Accuracy(test, predict):
    correction = 0
    for X in range(len(test)):
        if test[X][-1] is predict[X]:
            correction += 1
    return (correction/float(len(test)))*100
```

```python
test = [[4,2,3,'a'], [3,2,2,'a'], [5,3,6,'b']]
predict = ['a', 'b', 'a']
accuracy = Accuracy(test, predict)
print(accuracy)
```

```python
def mainFunction():
    training = []
    test = []
    split = 0.65
    loadingdataset('iris.txt', split, training, test)
    print('Train Set: ' + repr(len(training)))
    print('Test Set: ' + repr(len(test)))
    
    predict = []
    k = 3
    
    for X in range(len(test)):
        neighbors = Neighbors(training, test[X], k)
        res = Response(neighbors)
        predict.append(res)
        print('> predict=' + repr(res) + ', actual=' + repr(test[X][-1]))
    accuracy = Accuracy(test, predict)
    print('Accuracy: ' + repr(accuracy) + '%')
    
mainFunction()
```
